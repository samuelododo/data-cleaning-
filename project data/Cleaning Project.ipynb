## Multiple variables are stored in one column

```python
ebola = pd.read_csv('country_timeseries.csv')
ebola
```

```python
ebola_melted =  pd.melt(ebola, id_vars=['Date','Day'], var_name='Cases',
    value_name='value')
ebola_melted
```

```python
ebola_melted 
```

```python
ebola_melted[['stats', 'country']] = ebola_melted['Cases'].str.split('_', expand=True) 
```

```python
ebola_melted
```

## Variables are stored in both rows and columns 

```python
weather = pd.read_csv('weather.csv')
```

```python
weather 
```

```python
weather_melted = pd.melt(weather, id_vars=['id','year','month','element'],
    var_name='Day',
    value_name='temp',)
```

```python
weather_melted
```

```python
weather_tidy = pd.pivot_table(weather_melted,
    index=['id','year','month','Day'],
    columns='element',
    values='temp')
```

```python
weather_tidy.reset_index()
```
